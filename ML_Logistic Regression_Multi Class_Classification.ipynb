```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
dir(digits) #this dataset contains: data, images, target and target names
digits.data[0] #printing 0th row, nothing one image representaion in a row. one sample
plt.matshow(digits.images[0]) #matshow- numerical array acn be visualized as coloured image

plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])

#Cross check with data
for i in range(5):
    print(digits.target[i])
```

```python
#Create model and train logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

#Split data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(digits.data, digits.target, test_size=0.2)

#Train or fit the model
model.fit(X_train,y_train)

#Check Accuracy
print(model.score(X_test,y_test))

#Predict
print(model.predict(digits.data[0:4]))

#Confusion matrix
from sklearn import metrics
y_pred = model.predict(X_test)
conf_mat = metrics.confusion_matrix(y_test,y_pred)
print(conf_mat)

#Print accuracy, same as model score
print(metrics.accuracy_score(y_test,y_pred))

#Print confusion matrix report
print(metrics.classification_report(y_test,y_pred))

metrics.plot_confusion_matrix(model, X_test,y_test)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from math import*

#load data
df = pd.read_csv("C:/Prema/Data Science/ML algorithms/linear regression/homeprices.csv")

#Scatter plot to get an idea about the data whether it can be linearly separable or not
plt.xlabel('area')
plt.ylabel("price")
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()

plt.xlabel('age')
plt.ylabel("price")
plt.scatter(df.age,df.price,color='blue',marker='.')
plt.show()

plt.xlabel('bedrooms')
plt.ylabel("price")
plt.scatter(df.bedrooms,df.price,color='green',marker='*')
plt.show()
#from the plot, we can understand, it can be plot linearly

print(df.head())
#It is observed that there is one misssing value. Fill it with median or any other technique
#Fill with median
median_bedrooms = np.floor(df.bedrooms.median())
df.bedrooms = df.bedrooms.fillna(median_bedrooms) #Assign values in original column
print(df) # now the change is reflected in original data

#Train model with fit method
#reg =linear_model.LinearRegression(fit_intercept=True,normalize=False,copy_X=True,n_jobs=None)#default parameters,no need 2give
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)   #give as 2D, that is as data frame

#predict
print("\n",reg.predict([[3000,3,40]]),"\n")

# Checking co efficients of Model
print(reg.coef_)   #slope value
print(reg.intercept_)  #check the intercept value
print("\n",reg.score(df[['area','bedrooms','age']],df.price),"\n")  #if it close to 1, then it is good  model

#Checking all error values
from sklearn import metrics
#print('Mean Absolute Error:', metrics.mean_absolute_error([df.area,df.bedrooms,df.age],df.price))
#print('Mean Squared Error:', metrics.mean_squared_error(df.year,df.capita))
#print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(df.year,df.capita)))


```
